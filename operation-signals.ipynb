```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(-10,10,1000)
N = len(t)
A = 1
shift = 0
u = np.zeros(N)
for i in range(0,N):
    if t[i]>=-shift:
        u[i] = A
plt.grid()
plt.plot(t,u)
plt.show()
```

```python
# for negative values of shift, graph moves to right
t = np.linspace(-10,10,1000)
N = len(t)
A = 1
shift = -2
u = np.zeros(N)
for i in range(0,N):
    if t[i]>=-shift:
        u[i] = A
plt.grid()
plt.plot(t,u)
plt.show()
```

```python
# for positive values of shift, graph moves to left
t = np.linspace(-10,10,1000)
N = len(t)
A = 1
shift = 2
u = np.zeros(N)
for i in range(0,N):
    if t[i]>=-shift:
        u[i] = A
plt.grid()
plt.plot(t,u)
plt.show()
```
