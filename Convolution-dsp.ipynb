```python
import numpy as np
import matplotlib.pyplot as plt

x = eval(input("enter input values for x[n]: "))
nx = np.arange(0,len(x),1)
plt.subplot(2,2,1)
plt.stem(nx,x,'r')
plt.grid()

h = eval(input("enter impulse response: "))
nh = np.arange(0,len(h),1)
plt.subplot(2,2,2)
plt.grid()
plt.stem(nh,h,'r')

y = np.convolve(x,h)
ny = np.arange(0,len(y),1)
plt.subplot(2,2,3)
plt.stem(ny,y,'r')
plt.grid()
plt.show()
print(y,len(y))
```
