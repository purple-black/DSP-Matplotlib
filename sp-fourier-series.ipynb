```python
#question 
#  x[n] = cos((pi*n)/4)   -50<=n<=50


import matplotlib.pyplot as plt
import numpy as np

arr=[]
for i in range(-50,50,1):
    arr.append(np.cos((np.pi/4)*i))
    
    
x=[]
fs=[]
c=1
for sample in range(1, len(arr)-1):
    x.append(arr[sample-1])
    if arr[0] == arr[sample]:
        c+=1
        break
plt.stem(np.arange(-50,50),arr)
plt.show()

def summation(k):
    summ = 0
    for n in range(len(x)):
        e = np.exp(-2j * (np.pi) * k * n / len(x))
        summ = summ + float(x[n])*e
    return (summ)

for k in range(len(x)):
    fs.append(summation(k)/len(x))
    
ms = []
#ps = []
for sample in range(len(fs)):
    magn = np.abs((fs[sample].real**2 + fs[sample].imag**2)*0.5)
    ms.append(magn)
    #phase = np.arctan(fs[sample].imag/fs[sample].real)
    #ps.append(phase)

plt.stem(np.arange(-50, 48), ms)
plt.title("Magnitude spectrum")
plt.xlabel("k---->")
plt.ylabel("|X(k)|--->")
plt.show()
print(ms)
```
