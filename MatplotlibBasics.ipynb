```python
x = np.arange(0,5,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title("sin graph")
plt.show()
```

```python
x = np.arange(0,2*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title("sin graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)

plt.title("line graph")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,2,4,6,8,10,12])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['A','B','C']
values = [1,4,2]
plt.figure(dpi = 100)

bars = plt.bar(labels,values)
bars[0].set_hatch('*')
bars[1].set_hatch('/')
bars[2].set_hatch('O')
plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]
plt.figure(dpi = 100)

bars = plt.bar(labels,values)
patterns = ['o','/','*']

for bar in bars:
    bar.set_hatch(patterns.pop())

#bars[0].set_hatch('*')
#bars[1].set_hatch('/')
#bars[2].set_hatch('O')
plt.show()
```
